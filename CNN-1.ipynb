```python
from tensorflow import keras
from keras.datasets import mnist

(X_train , y_train) , (X_test , y_test) = mnist.load_data()

X_train.shape
```

```python
X_train[5]
```

```python
import matplotlib.pyplot as plt

plt.imshow(X_train[5] , cmap="gray")
plt.show()
```

# Pre Processing

```python
X_train = X_train.reshape((-1, 28, 28 , 1)).astype("float32") / 255.0
X_test = X_test.reshape((-1, 28, 28 , 1)).astype("float32") / 255.0

X_train.shape
```

# Model Definition

```python
import tensorflow as tf

model = keras.Sequential()
model.add(tf.keras.Input(shape=(28,28,1)))
model.add(keras.layers.Conv2D(filters=32, kernel_size=(3,3) , strides=(1,1) , padding="valid", activation="relu" ))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(units=128 , activation="relu"))
model.add(keras.layers.Dense(units=10 , activation="softmax"))


model.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.sparse_categorical_crossentropy , metrics=["accuracy"])

model.summary()
```

```python
hist = model.fit(X_train , y_train , epochs=8 , batch_size=256  , validation_data=(X_test, y_test), verbose=1)

```

```python
plt.plot(hist.history["accuracy"] , color="red")
plt.plot(hist.history["val_accuracy"] , color="blue")

plt.show()
```
