Different Operation

```python
a = tf.constant([3,5])
b = tf.constant([1,4])
```

```python
#sum
c = tf.add(a,b)
c.numpy()
```

```python
# subtract
d= tf.subtract(a,b)
d.numpy()
```

```python
#multiplya
a_1 = tf.constant([[2,3],[0,5]])
b_1 = tf.constant([[1,3],[6,4]])

ew = tf.multiply(a_1,b_1)
ew
```

```python
mm = tf.matmul(a_1, b_1)
mm
```

```python
#divide
div = tf.divide(a_1,b_1)
div
```

```python
# broadcasting
a_2 = tf.constant([1,2,3])
b_2 = tf.constant(4) # 4 => [4,4,4]

c_2 = a_2 * b_2
c_2
```

# Min , Max, Mean ,STD

```python
a = tf.constant([[3,4],[2,5]], dtype=tf.float32)
a
a.numpy()
```

```python
min = tf.reduce_min(a, axis=1)
min
```

```python
#numpy
a3= np.array([[3,4],[2,5]])
a3.mean()
```

```python
avg = tf.reduce_mean(a)
avg.numpy()
```

```python
avg = tf.reduce_mean(a, axis=0)
avg.numpy()
```

```python
avg = tf.reduce_mean(a, axis=1)
avg.numpy()
```

```python
std = tf.math.reduce_std(a)
std.numpy()
```
