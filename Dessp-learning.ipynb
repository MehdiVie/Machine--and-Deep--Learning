```python
import tensorflow as tf
from tensorflow import keras
```

# Data Definition

```python
# y = x * 3 - 5
x = tf.constant([-1 , -3 , 7 , 4 , 2 , 0] , dtype=tf.float32)
y = x * 3 - 5
```

```python
x.numpy()
```

```python
y.numpy()
```

# Model Definition

```python
model = keras.Sequential([
    keras.layers.Dense(units=1 , input_shape=(1,) , activation=None)
])
```

```python
# (y - y_pred) ** 2
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
model.summary()
```

```python
model.fit(x, y, batch_size=1, epochs=500)
```

# Test

```python
x_test = tf.constant([-10,0.5,1.5] , dtype=tf.float32)
y_test = x_test * 3 -5
```

```python
x_test.numpy() , y_test.numpy()
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_test.numpy() , y_test.numpy() , color="red")
plt.scatter(x_test.numpy() , y_pred , color="blue")
plt.show()
```
