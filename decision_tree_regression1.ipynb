# Decision Tree Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('diabetes1.csv', delimiter=';')
dataset.to_csv('rando2.csv', index=False)
X = dataset.iloc[:, 0:-1].values
y = dataset.iloc[:, -1].values
```

```
X.shape
```

## Training the Decision Tree Regression model on the whole dataset

```
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

## Predicting a new result

```
regressor.predict([[7, 148, 75, 40, 80, 35, 0.652, 55]])
```

## Visualising the Decision Tree Regression results (higher resolution)

```
X_grid = np.zeros((len(np.arange(np.min(X[:, 7]), np.max(X[:, 7]), 0.01)), X.shape[1]))
X_grid.shape
```

```
X_grid[:, 7] = np.arange(np.min(X[:, 7]), np.max(X[:, 7]), 0.01)
X_grid[:, 7].shape
```

```
a = range(X.shape[1])
a
```

```
X_age_plotting = X[:, 7]
X_age_plotting.shape
```

```
X_grid = X_grid.reshape((len(X_grid), X.shape[1]))
X_grid.shape
```

```
plt.scatter(X_age_plotting, y, color = 'red')
plt.plot(X_grid[:, 7], regressor.predict(X_grid), color = 'blue') # Use only age (7th column) for plotting on X-axis


plt.scatter(X_age_plotting, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision_Tree)')
plt.xlabel('Age')
plt.ylabel('Diabet')
plt.show()
```
